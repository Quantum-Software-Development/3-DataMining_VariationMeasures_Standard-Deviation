```
# Load the dataset
df = pd.read_csv('/content/banco.csv', sep=';')

# Display the first few rows of the DataFrame
display(df.head())

# Display information about the DataFrame
display(df.info())
```

```
# Identify numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Calculate mean, median, and mode for numerical columns
descriptive_stats = pd.DataFrame({
    'mean': df[numerical_cols].mean().round(4),
    'median': df[numerical_cols].median(),
    'mode': df[numerical_cols].mode().iloc[0] # Mode can have multiple values, taking the first one
})

# Display the descriptive statistics
print("Descriptive Statistics for Numerical Columns:")
display(descriptive_stats)
```

```
# Set the style for the plots
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.facecolor'] = 'black'
plt.rcParams['axes.facecolor'] = 'black'
plt.rcParams['text.color'] = 'white'
plt.rcParams['axes.labelcolor'] = 'white'
plt.rcParams['xtick.color'] = 'white'
plt.rcParams['ytick.color'] = 'white'

# Visualize numerical columns with histograms
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x=col, color='turquoise', kde=True)
    plt.title(f'Distribution of {col}', color='white')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Visualize numerical columns with box plots
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, y=col, color='turquoise')
    plt.title(f'Box Plot of {col}', color='white')
    plt.ylabel(col)
    plt.show()
```
